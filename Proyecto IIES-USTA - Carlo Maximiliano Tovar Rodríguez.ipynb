```python
board=[[3*j+i+1 for i in range(3)]for j in range(3)]
print(board)

def DisplayBoard(board):
    print("+-------"*3,"+", sep="")
    for fila in range(3):
        print("|       "*3, "|", sep="")
        for columna in range(3):        
            print("|   " + str(board[fila][columna]),"  ", end="")
        print("|")
        print("|       "*3,"|", sep="")
        print("+-------"*3, "+", sep="")
DisplayBoard(board)
# la función acepta un parámetro el cual contiene el estado actual del tablero
# y lo muestra en la consola
#
```

```python
def EnterMove(board):
    ok = False
    
    while not ok:
        movimiento = input("Ingresa tu movimiento: ")
        ok = len(movimiento)==1 and movimiento >='1' and movimiento <='9' # Valido
        if not ok:
            print("Movimiento erroneo, intente nuevamente")
            continue
        movimiento = int(movimiento)-1
        fila = movimiento//3 #Fila
        columna = movimiento %3 #Columna
        asig = board[fila][columna]
        ok = asig not in ['X','O']
        if not ok: #Verdadero
            print("!El campo esta ocupado, intenta nuevamente!") #Validar campo libre
            continue
        board[fila][columna] = 'O' #Actualiza movimiento del usuario    
    return board
```

```python
def MakeListOfFreeFields(board):
    libre=[]
    for fila in range(3):
        for columna in range(3):
            if board[fila][columna] not in ['X','O']:
                libre.append((fila, columna))
    return libre
# la función examina el tablero y construye una lista de todos los cuadros vacíos 
# la lista esta compuesta por tuplas, cada tupla es un par de números que indican la fila y columna
#
```

```python
def VictoryFor(board, asig):
    if asig =='X': #Buscando X
        quien='Yo'
    elif asig =='O':
        quien='Tu'
    else:
        quien=None
        
    diag1 = diag2 =True
    
    for i in range(3):
        if board[i][0] == asig and board[i][1] == asig and board[i][2] == asig: #Ganó por fila 
            return quien
        if board[0][i] == asig and board[1][i] == asig and board[2][i] == asig: #Ganó por columna
            return quien
        if board[i][i] != asig: #Ganó por diagonal
            diag1 = False
        if board[i][2-i] != asig: #Ganó por diagonal
            diag2 = False
    if diag1 or diag2:
        return quien
    
    return None
        
#
# la función analiza el estatus del tablero para verificar si
# el jugador que utiliza las 'O's o las 'X's ha ganado el juego
#
```

```python
def DrawMove(board):
    libre = MakeListOfFreeFields(board) #Lista de espacios libres
    cont = len(libre)
    if cont > 0: #Validar que existan campos libres
        cual = randrange(cont)
        fila, columna = libre[cual]
        board[fila][columna] = 'X'
#
# la función dibuja el movimiento de la maquina y actualiza el tablero
#
```

```python
## board=[[3*j+i+1 for i in range(3)]for j in range(3)]
#print(board)
board[1][1]='X' #Primera asignación.
libre = MakeListOfFreeFields(board)
turno_usr=True

while len(libre):
    DisplayBoard(board)
    if turno_usr:
        EnterMove(board)
        win=VictoryFor(board, 'O')
    else:
        DrawMove(board)
        win=VictoryFor(board, 'X')
        
    
    if win != None:
        break
    turno_usr = not turno_usr
    libre = MakeListOfFreeFields(board)

DisplayBoard(board)
if win == 'Tu':
    print("!Has ganado!")
elif win == 'Yo':
    print("!He ganado!")
else:
    print("!Empate!")

        
    
#def EnterMove(board):
#    while 
#
# la función acepta el estado actual del tablero y pregunta al usuario acerca de su movimiento, 
# verifica la entrada y actualiza el tablero acorde a la decisión del usuario
#
```
